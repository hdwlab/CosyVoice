```python
from huggingface_hub import snapshot_download
snapshot_download('FunAudioLLM/CosyVoice-300M', local_dir='pretrained_models/CosyVoice-300M')
snapshot_download('FunAudioLLM/Fun-CosyVoice3-0.5B-2512', local_dir='pretrained_models/Fun-CosyVoice3-0.5B')
```

```python
import torchaudio
from cosyvoice.cli.cosyvoice import AutoModel

# <|endofprompt|> is needed for 0.5B model
# 1. zero-shot TTS (source text required)
# 2. cross-lingual TTS (source text not required)
cosyvoice = AutoModel(model_dir='pretrained_models/Fun-CosyVoice3-0.5B')

for i, j in enumerate(
        cosyvoice.inference_zero_shot(
            'あのイーハトーヴォのすきとおった風', 
            'そして、５レーンがアルフレッド<|endofprompt|>',  
            './alfred.flac', stream=False
        )
    ):
    torchaudio.save(f'./zero_shot_{i}.wav', j["tts_speech"], cosyvoice.sample_rate)

for i, j in enumerate(cosyvoice.inference_cross_lingual(
    'あのイーハトーヴォのすきとおった風<|endofprompt|>', 
    './sample_01.wav'
)):
    torchaudio.save(f'./cross_lingual_{i}.wav', j['tts_speech'], cosyvoice.sample_rate)
```

```python
import torchaudio
from cosyvoice.cli.cosyvoice import AutoModel

cosyvoice = AutoModel(model_dir='pretrained_models/CosyVoice-300M')
# 1. zero-shot TTS (source text required)
# 2. cross-lingual TTS (source text not required)
# 3. voice conversion (source text not required)

for i, j in enumerate(
        cosyvoice.inference_zero_shot(
            'あのイーハトーヴォのすきとおった風', 
            'そして、５レーンがアルフレッド',
            './alfred.flac', stream=False
        )
    ):
    torchaudio.save(f'./zero_shot_{i+1}.wav', j["tts_speech"], cosyvoice.sample_rate)

for i, j in enumerate(cosyvoice.inference_cross_lingual(
    'あのイーハトーヴォのすきとおった風', 
    './sample_01.wav'
)):
    torchaudio.save(f'./cross_lingual_{i+1}.wav', j['tts_speech'], cosyvoice.sample_rate)

for i, j in enumerate(cosyvoice.inference_vc('./sample_01.wav', './alfred.flac')):
    torchaudio.save(f'./vc_{i+1}.wav', j['tts_speech'], cosyvoice.sample_rate)
```
