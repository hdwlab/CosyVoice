```python
from huggingface_hub import snapshot_download

model_dir_300m = snapshot_download('FunAudioLLM/CosyVoice-300M', revision="24c40509c3c5ea6fe06b5f8790ff99e3714a6bee")
model_dir_500m = snapshot_download('FunAudioLLM/Fun-CosyVoice3-0.5B-2512', revision="29e01c4e8d000f4bcd70751be16fa94bf3d85a18")
```

```python
from cosyvoice.cli.cosyvoice import AutoModel

cosyvoice = AutoModel(model_dir=model_dir_300m, device="cpu")
type(cosyvoice).__name__
```

```python
import torchaudio
from cosyvoice.cli.cosyvoice import AutoModel

# <|endofprompt|> is needed for 0.5B model
# 1. zero-shot TTS (source text required)
# 2. "cross-lingual" TTS (source text not required)
cosyvoice = AutoModel(model_dir=model_dir_500m)

for i, j in enumerate(
        cosyvoice.inference_zero_shot(
            'あのイーハトーヴォのすきとおった風',
            'そして、５レーンがアルフレッド<|endofprompt|>',
            './alfred.flac', stream=False
        )
    ):
    torchaudio.save(f'./zero_shot_{i}.wav', j["tts_speech"], cosyvoice.sample_rate)

for i, j in enumerate(cosyvoice.inference_cross_lingual(
    'あのイーハトーヴォのすきとおった風<|endofprompt|>',
    './sample_01.wav'
)):
    torchaudio.save(f'./cross_lingual_{i}.wav', j['tts_speech'], cosyvoice.sample_rate)
```
