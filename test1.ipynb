```python
import os
import sys

# os.environ["CUDA_VISIBLE_DEVICES"] = ""
sys.path.append('third_party/Matcha-TTS')
```

```python
from cosyvoice.cli.cosyvoice import AutoModel

cosyvoice = AutoModel(model_dir='pretrained_models/CosyVoice-300M-SFT')
```

```python
print(cosyvoice.list_available_spks())
```

```python
import torchaudio

# text = '我思うが故に我あり'
# text = 'あのイーハトーヴォのすきとおった風'
# text = 'アメンボ赤いなあいうえお'
# text = 'こんにちは'
text = '結合することができないから矛盾するというのである。'
for i, j in enumerate(cosyvoice.inference_sft(text, '日语男', stream=False)):
    torchaudio.save(f'sft_{text}.wav', j['tts_speech'], cosyvoice.sample_rate)
```
